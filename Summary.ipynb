```python
import pandas as pd
import seaborn as sns
from matplotlib.font_manager import FontProperties
```

```python
%matplotlib inline
```

```python
sns.set(color_codes=True)
sns.set(font=['Meiryo'])
```

```python
data1 = pd.read_csv("~/Downloads/NHK_Summary.txt", header=None)
data1.columns =["Date", "Word", "Count", "Provided_by"]
data1["Date"]  = pd.to_datetime(data1["Date"])

data2 = pd.read_csv("~/Downloads/TBS_Summary.txt", header=None)
data2.columns =["Date", "Word", "Count", "Provided_by"]
data2["Date"]  = pd.to_datetime(data2["Date"])

data = pd.concat([data1, data2])
```

```python
topwords = data.groupby("Word")["Count"].sum().sort_values(ascending=False).index
data.groupby("Word")["Count"].sum().sort_values(ascending=False)[ :3]
```

```python
def f7(seq):
    seen = set()
    seen_add = seen.add
    return [x for x in seq if not (x in seen or seen_add(x))]
```

```python
top = data[ data["Word"].isin([topwords[4], topwords[7], topwords[0]])].sort_values(by="Date")
labels = [str(i+1) for i in range(32)]
fig1 = sns.pointplot(x="Date", y="Count", data=top.sort_values(by="Date"), hue="Word", markers=["o", "x", "s"], linestyles=["-", "--", "-."],
                    scale = 0.8, ci=None)
fig1 = fig1.set_xticklabels(f7(top["Date"].dt.date), rotation=90)
```

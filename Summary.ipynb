```python
import pandas as pd
import seaborn as sns
from matplotlib.font_manager import FontProperties
```

```python
%matplotlib inline
```

```python
sns.set(color_codes=True)
sns.set(font=['Meiryo'])
```

```python
data = pd.read_csv("~/Downloads/Summary.txt", header=None)
data.columns =["Date", "Word", "Count", "Provided_by"]
data["Date"]  = pd.to_datetime(data["Date"])
```

```python
topwords = data.groupby("Word")["Count"].sum().sort_values(ascending=False).index
data.groupby("Word")["Count"].sum().sort_values(ascending=False)[:5]
```

```python
top = data[ data["Word"].isin([topwords[4], topwords[7], topwords[0]])] 
labels = [str(i+1) for i in range(32)]
fig1 = sns.pointplot(x="Date", y="Count", data=top.sort_values(by="Date"), hue="Word", markers=["o", "x", "s"], linestyles=["-", "--", "-."],
                    scale = 0.8)
fig1 = fig1.set_xticklabels(labels, rotation=0)
```

```python
top = data[ data["Word"].isin(["日本", "アメリカ", "ヨルダン"])]
labels = [str(i+1) for i in range(32)]
fig2 = sns.pointplot(x="Date", y="Count", data=top.sort_values(by="Date"), hue="Word", markers=["o", "x", "D"], linestyles=["-", "--", "-."],
                    scale = 0.8)
fig2 = fig2.set_xticklabels(labels, rotation=0)
```
